```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
```

```python
path = '/content/drive/MyDrive/roo_data.csv'
df = pd.read_csv(path)
```

```python
df['can work long time before system?'] = le.fit_transform(df['can work long time before system?'])
```

```python
df['self-learning capability?'] = le.fit_transform(df['self-learning capability?'])
df['Extra-courses did'] = le.fit_transform(df['Extra-courses did'])
df['certifications'] = le.fit_transform(df['certifications'])
df['workshops'] = le.fit_transform(df['workshops'])
df['talenttests taken?'] = le.fit_transform(df['talenttests taken?'])
df['olympiads'] = le.fit_transform(df['olympiads'])
df['reading and writing skills'] = le.fit_transform(df['reading and writing skills'])
df['memory capability score'] = le.fit_transform(df['memory capability score'])
df['Interested subjects'] = le.fit_transform(df['Interested subjects'])
df['interested career area '] = le.fit_transform(df['interested career area '])
df['Job/Higher Studies?'] = le.fit_transform(df['Job/Higher Studies?'])
df['Type of company want to settle in?'] = le.fit_transform(df['Type of company want to settle in?'])
df['Taken inputs from seniors or elders'] = le.fit_transform(df['Taken inputs from seniors or elders'])
df['interested in games'] = le.fit_transform(df['interested in games'])
df['Interested Type of Books'] = le.fit_transform(df['Interested Type of Books'])
df['Salary Range Expected'] = le.fit_transform(df['Salary Range Expected'])
df['In a Realtionship?'] = le.fit_transform(df['In a Realtionship?'])
df['Gentle or Tuff behaviour?'] = le.fit_transform(df['Gentle or Tuff behaviour?'])
df['Management or Technical'] = le.fit_transform(df['Management or Technical'])
df['Salary/work'] = le.fit_transform(df['Salary/work'])
df['hard/smart worker'] = le.fit_transform(df['hard/smart worker'])
df['worked in teams ever?'] = le.fit_transform(df['worked in teams ever?'])
df['Introvert'] = le.fit_transform(df['Introvert'])
df['Suggested Job Role'] = le.fit_transform(df['Suggested Job Role'])
```

Seperating x and y 

```python
y = df['Suggested Job Role']
```

```python
y
```

```python
x=df.drop('Suggested Job Role',axis=1)
```

```python
x
```

Training using ANN

Splitting the data into 90-10 ratio.

```python
x_training,x_testdata,y_training,y_testdata=train_test_split(x,y,test_size=0.1)
clf10 = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(25,), random_state=1)
clf10.fit(x_training, y_training)
p10=clf10.predict(x_testdata)
print(classification_report(y_testdata, p10))
print(confusion_matrix(y_testdata, p10))
accuracy10 = clf10.score(x_testdata, y_testdata)
accuracy10
```

Splitting the data into 80-20 ratio

```python
x_training,x_testdata,y_training,y_testdata=train_test_split(x,y,test_size=0.2)
clf20 = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(25,), random_state=1)
clf20.fit(x_training, y_training)
p20 = clf20.predict(x_testdata)
print(classification_report(y_testdata, p20))
print(confusion_matrix(y_testdata, p20))
accuracy20 = clf20.score(x_testdata, y_testdata)
accuracy20
```

Splitting the data into 70-30 ratio.

```python
x_training,x_testdata,y_training,y_testdata=train_test_split(x,y,test_size=0.3)
clf30 = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(25,), random_state=1)
clf30.fit(x_training, y_training)
p30=clf30.predict(x_testdata)
print(classification_report(y_testdata, p30))
print(confusion_matrix(y_testdata, p30))
accuracy30 = clf30.score(x_testdata, y_testdata)
accuracy30
```

Splitting the data into 60-40 ratio.

```python
x_training,x_testdata,y_training,y_testdata=train_test_split(x,y,test_size=0.4)
clf40 = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(25,), random_state=1)
clf40.fit(x_training, y_training)
p40=clf40.predict(x_testdata)
print(classification_report(y_testdata, p40))
print(confusion_matrix(y_testdata, p40))
accuracy40 = clf40.score(x_testdata, y_testdata)
accuracy40
```

Splitting the data into 50-50 ratio.

```python
x_training,x_testdata,y_training,y_testdata=train_test_split(x,y,test_size=0.5)
clf50 = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(25,), random_state=1)
clf50.fit(x_training, y_training)
p50=clf50.predict(x_testdata)
print(classification_report(y_testdata, p50))
print(confusion_matrix(y_testdata, p50))
accuracy50 = clf50.score(x_testdata, y_testdata)
accuracy50
```

```python
accuracy = []
accuracy.append(accuracy10)
accuracy.append(accuracy20)
accuracy.append(accuracy30)
accuracy.append(accuracy40)
accuracy.append(accuracy50)

split_ratio_percentage = []
split_ratio_percentage.append('90-10')
split_ratio_percentage.append('80-20')
split_ratio_percentage.append('70-30')
split_ratio_percentage.append('60-40')
split_ratio_percentage.append('50-50')
```

```python
table = pd.DataFrame({'Split Ratio Percentage': split_ratio_percentage, 'Accuracy': accuracy})
table
```

```python
fig = plt.subplots(figsize =(5, 5))

plt.plot(split_ratio_percentage, accuracy)

plt.xlabel("split_ratio_percentage")
plt.ylabel('accuracy')
plt.title("Accuracy w.r.t. Split Ratio")

plt.legend()

plt.show()
```
